# Three Dimensional Force Systems

## Bedford, Fowler. Statics, problem 3.74

If the mass of the bar *AB* is negligible compared to the mass of the suspended object *E*, the bar exerts a force on the ball at *B* that points from *A* toward *B*. The mass of the object *E* is $ 200 \; \mathrm{kg} $. The *y* axis points upward.Determine the tensions in the cables *BC* and *BD*.

*Strategy*: Draw a free-body diagram of the ball at *B*. (The weight of the ball is negligible.)

bf-p3_74-small.png

### Strategy


We isolate the ball *B* and obtain a free-body diagram subjected to forces due to the tensions in the cables and the bar. Because the sums of the external forces in the *x*, *y*, and *z* directions must each equal zero, we can obtain three equilibrium equations for the three unknown forces. To do so, we must express the forces exerted by the cables and the bar in terms of their components.

### Solution

**Draw the Free-Body Diagram and Apply the Equilibrium of Forces**

 The sum of the forces must equal zero:

$$ \Sigma \, \mathbf{F} = \mathbf{F}_{AB} + \mathbf{T}_{BC} + \mathbf{T}_{BD} + \mathbf{w} = 0 $$

**Write the Forces in Terms of Their Components**

Here, $ \mathbf{w} $ represents the weight of the suspended object, with a magnitude $ w = m \cdot g = 200 \, \mathrm{kg} \cdot 9,8 \, \mathrm{^m/_{s^2}} = 1960 \, \mathrm{N} $. Hence, the vector that represents the weight is:

$$ \mathbf{w} = -w \, \mathbf{\hat{j}} $$


To express the force $ \mathbf{F}_{AB} $ in terms of its components it should be written as the product of the magnitude of the force $ F_{AB} $ by the bar *AB* and the unit vector $ \mathbf{\hat{e}}_{AB} $:

$$ \mathbf{F}_{AB} = F_{AB} \cdot \mathbf{\hat{e}}_{AB} $$

A unit vector that has the same direction as the force $ \mathbf{F}_{AB} $ is built dividing the position vector $ \mathbf{r}_{AB} $ from point *A* to point *B* by its magnitude. Remember that, as stated in the problem, the bar *AB* exerts a force on the ball at *B* that points from *A* toward *B*.

$$ \mathbf{\hat{e}}_{AB} = \frac{\mathbf{r}_{AB}}{r_{AB}} $$

$$ \mathbf{r}_{AB} = \mathbf{r}_{B} - \mathbf{r}_{A} $$

As *A* is the origin of the coordinate system:

$$ \mathbf{r}_{AB} = \left( 4 \, \mathbf{\hat{i}} + 3 \, \mathbf{\hat{j}} + 1 \, \mathbf{\hat{k}} \right) \, \mathrm{m} $$

While its magnitude is:

$$ r_{AB} = \sqrt{4^2 + 3^2 + 1^2} \, \mathrm{m} = \sqrt{26} \, \mathrm{m} $$

$$ \mathbf{\hat{e}}_{AB} = \frac{\mathbf{r}_{AB}}{r_{AB}} = \frac{1}{\sqrt{26}} \left( 4 \, \mathbf{\hat{i}} + 3 \, \mathbf{\hat{j}} + 1 \, \mathbf{\hat{k}} \right) $$

In this way, $ \mathbf{F}_{AB} $ is finally expressed as:

$$ \mathbf{F}_{AB} = F_{AB} \cdot \frac{1}{\sqrt{26}} \left( 4 \, \mathbf{\hat{i}} + 3 \, \mathbf{\hat{j}} + 1 \, \mathbf{\hat{k}} \right) $$

Using the same procedure the forces $ \mathbf{T}_{BC} $ and $ \mathbf{T}_{BD} $ can be expressed in terms of their components.

For $ \mathbf{T}_{BC} $:

$$ \mathbf{T}_{BC} = T_{BC} \cdot \mathbf{\hat{e}}_{BC} $$

$$ \mathbf{\hat{e}}_{BC} = \frac{\mathbf{r}_{BC}}{r_{BC}} $$

$$ \mathbf{r}_{BC} = \mathbf{r}_{C} - \mathbf{r}_{B} $$

$$ \mathbf{r}_{BC} = \left( 4 \, \mathbf{\hat{j}} -3 \, \mathbf{\hat{k}} \right) \, \mathrm{m} - \left( 4 \, \mathbf{\hat{i}} + 3 \, \mathbf{\hat{j}} + 1 \, \mathbf{\hat{k}} \right) \, \mathrm{m} $$

$$ \mathbf{r}_{BC} = \left( -4 \, \mathbf{\hat{i}} + 1 \, \mathbf{\hat{j}} -4 \, \mathbf{\hat{k}} \right) \, \mathrm{m} $$

$$ r_{BC} = \sqrt{(-4)^2 + 1^2 + (-4)^2} \, \mathrm{m} = \sqrt{33} \, \mathrm{m} $$

$$ \mathbf{T}_{BC} = T_{BC} \cdot \frac{1}{\sqrt{33}} \left( -4 \, \mathbf{\hat{i}} + 1 \, \mathbf{\hat{j}} - 4 \, \mathbf{\hat{k}} \right) $$

In the case of $ \mathbf{T}_{BD} $:

$$ \mathbf{T}_{BD} = T_{BD} \cdot \mathbf{\hat{e}}_{BD} $$

$$ \mathbf{\hat{e}}_{BD} = \frac{\mathbf{r}_{BD}}{r_{BD}} $$

$$ \mathbf{r}_{BD} = \mathbf{r}_{D} - \mathbf{r}_{B} $$

$$ \mathbf{r}_{BD} = \left( 5 \, \mathbf{\hat{j}} + 5 \, \mathbf{\hat{k}} \right) \, \mathrm{m} - \left( 4 \, \mathbf{\hat{i}} + 3 \, \mathbf{\hat{j}} + 1 \, \mathbf{\hat{k}} \right) \, \mathrm{m} $$

$$ \mathbf{r}_{BD} = \left( -4 \, \mathbf{\hat{i}} + 2 \, \mathbf{\hat{j}} + 4 \, \mathbf{\hat{k}} \right) \, \mathrm{m} $$

$$ r_{BD} = \sqrt{4^2 + 2^2 + 4^2} \, \mathrm{m} = 6 \, \mathrm{m} $$

$$ \mathbf{T}_{BD} = T_{BD} \cdot \frac{1}{3} \left( -2 \, \mathbf{\hat{i}} + 1 \, \mathbf{\hat{j}} + 2 \, \mathbf{\hat{k}} \right) $$

**Substitute the Forces into the Equilibrium Equation**

$$ \Sigma \, \mathbf{F} = \mathbf{F}_{AB} + \mathbf{T}_{BC} + \mathbf{T}_{BD} + \mathbf{w} = 0 $$

Because the $\mathbf{\hat{i}}, \, \mathbf{\hat{j}}$, and $\mathbf{\hat{k}}$ components must each equal zero, this results in three equilibrium equations:

$$ \begin{matrix}
\Sigma \, F_x: & \frac{4}{\sqrt{26}} \, F_{AB} &  - \frac{4}{\sqrt{33}} \, T_{BC} & - \frac{2}{3} \, T_{BD} & = 0 \\
\Sigma \, F_y: & \frac{3}{\sqrt{26}} \, F_{AB} & + \frac{1}{\sqrt{33}} \, T_{BC} & + \frac{1}{3}  \, T_{BD} & = w \\
\Sigma \, F_z: & \frac{1}{\sqrt{26}} \, F_{AB} & - \frac{4}{\sqrt{33}} \, T_{BC} & + \frac{2}{3} \, T_{BD} & = 0
\end{matrix} $$

Notice that in $ \Sigma \, F_y $, the weight $w$ is at the right side of the equation. Using the *Solutions* command in *GeoGebra CAS Calculator* to solve these three equations yields:

bf-p3_74-solutions.png

$$ F_{AB} = \frac{8 \, \sqrt{26}}{35} \, w = 2284,4 \, \mathrm{N} $$

$$ T_{BC} = \frac{\sqrt{33}}{7} \, w = 1608,5 \, \mathrm{N} $$

$$ T_{BD} = \frac{18}{35} \, w = 1008 \, \mathrm{N} $$
